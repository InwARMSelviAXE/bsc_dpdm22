```python
from sklearn import datasets
iris_data = datasets.load_iris()
```

```python
X = iris_data['data']
y = iris_data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold 
```

```python
model = DecisionTreeClassifier()

kf = KFold(n_splits=10, shuffle=True, random_state=42)

scores = cross_val_score(model, X, y, cv=kf)

print('Accuracy: %0.2f (+/-%0.2f)' % (scores.mean(), scores.std() * 2))
```

```python
model = DecisionTreeClassifier(max_depth=2)

kf = KFold(n_splits=10, shuffle=True, random_state=42)

scores = cross_val_score(model, X, y, cv=kf)

print('Accuracy: %0.2f (+/-%0.2f)' % (scores.mean(), scores.std() * 2))
```

#HW6 วนลูปทำ 10-fold cross validuation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
shuffled_df = shuffled_df.sort_values('fold')
shuffled_df
```

```python
for i in range(1, 11):
    FD = shuffled_df[shuffled_df['fold'] == i]
    print(f"fold {i}:")
    print(FD)
   
    model = DecisionTreeClassifier()

    # Define the k-fold cross validation
    kf = KFold(n_splits=10, shuffle=True, random_state=6)

    # Select the features and labels for this fold
    X_fold = FD.drop(['label'], axis=1)
    y_fold = FD['label']

    # Train and evaluate the model using 10-fold cross validation
    scores = cross_val_score(model, X_fold, y_fold, cv=kf)

    # Print the average accuracy and standard deviation across all folds
    print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
