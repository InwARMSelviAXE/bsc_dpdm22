```python
def check(in1,in2) :
    return max(in1/in2,in2/in1)
```

```python
check(100,12)
```

```python
check(100,1200)
```

```python
check1 = list()
for row in house1.iterrows():
  check1.append(check(row[1]['ราคาทรัพย์'],row[1]['ประเมิน1']))
check1
```

```python
check2 = list()
for row in house1.iterrows():
  check2.append(check(row[1]['ราคาทรัพย์'],row[1]['ประเมิน2']))
check2
```

```python
house1['check1'] = check1
house1['check2'] = check2
house1
```

```python
house1.sort_values(by=['check1'],ascending=False)
```
